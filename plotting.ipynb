### Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
len(t1), len(t2)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
```

```python
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'ro')
plt.show()
```

### Adjusting the spacing around subplots

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
#plt.subplots_adjust(wspace=, hspace=0)
plt.show()
```

### Ticks, Labels, and Legends

***Working with Ticks***

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())

ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two', 'three', 'four', 'five'], rotation = 30, fontsize='small', color='r')

ax.set_xlabel('Stages')
ax.set_ylabel('Y-Label')

ax.set_title('My first matplotlib plot')
#ax.set_xticklabels?
```

***Legends example 1 ***

```python
data = randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
#plt.legend(loc=5)
```

***Legends example 2***

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k', label='one')
ax.plot(randn(1000).cumsum(),'k--', label='two')
ax.plot(randn(1000).cumsum(),'k.', label='three')

ax.legend(loc='best')
## Or plt.legend(loc='best')
```

### Saving Plots to File

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k', label='one')
ax.legend(loc='best')
plt.savefig('savefig.png')
```

***Show the percentage of data points for each party size on each day. ***

```python
tips = pd.read_csv('/home/nitin/Documents/ebooks/Onlinecourses/Python/pydata-book-master/ch08/tips.csv')
tips.head()
```

```python
## Create crosstab
party_counts = pd.crosstab(tips['day'],tips['size'])
party_counts
```

```python
## Normalize to sum to 1
party_pcts = party_counts.div(party_counts.sum(1).astype(float), axis=0)

## Plot it.
party_pcts.plot(kind='bar', stacked=True)
```

**Plotting histogram and density plot together**

```python
comp1 = np.random.normal(0, 1, size=200)  # N(0, 1)
comp2 = np.random.normal(10, 2, size=200) # N(10, 4)

print(len(comp1))
print(len(comp2))
```

```python
values = Series(np.concatenate([comp1,comp2]))
values.hist(bins = 100,color='k', normed=True)
values.plot(kind='kde',color='r')
```

### Scatter Plots

**Scatter plots are useful for examining the relationship between two one-dimensional data series.**

```python
macro = pd.read_csv('/home/nitin/Documents/ebooks/Onlinecourses/Python/pydata-book-master/ch08/macrodata.csv')
data = macro[['cpi','m1','tbilrate','unemp']]
data.head()
```

```python
print(type(np.log(data)))
trans_data = np.log(data)
trans_data = trans_data.diff()
trans_data = trans_data.dropna()
trans_data[-5:]
```

```python
trans_data.plot(kind = 'scatter', x = 'm1', y = 'unemp',title = 'Changes in log %s vs. log %s' % ('m1', 'unemp'))
```

### Scatter Matrix

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(trans_data, alpha=0.2, diagonal='kde')
```
